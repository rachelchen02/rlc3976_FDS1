```python
import RC_stats_module
user_inp = 0
numbers = []
#Menu
while user_inp != 8: 
    print("Statistical Calculator\n1. Input Numbers\n2. Mean\n3. Variance\n4. Standard Deviation\n5. Standard Error\n6. Z-score of a new observation\n7. Summary of the statistics\n8. Quit")
    user_inp = input("Input the number of the operation: ")
    #Error case where someone types letters
    if user_inp.isalpha() == True:
        print("This is not a number. Try again!\n")
    else:
        user_inp = int(user_inp)
        #Quit
        if user_inp == 8:
            break
        #Input numbers
        elif user_inp == 1:
            numbers = []
            while True:
                number = input("Input a series of numbers one by one. Type 'done' when done:")
                #User is done inputting numbers
                if number == 'done':
                    print("\n")
                    break
                #Error case where someone types letters 
                elif number.isalpha() == True:
                    print("This is not a number. Try again!\n")
                #Adding user inputs to a list
                else: 
                    number = int(number)
                    numbers.append(number)
        #Error where someone types a number that's not the operation 
        elif user_inp >= 9 or user_inp <= 0 :
            print("This is not one of the operations. Try again!\n")        
        #Error when someone tries an operation w/o inputting numbers
        elif user_inp != 1 and numbers == []:
            print("There are no numbers inputted. Try again!\n")
        #Mean calculation
        elif user_inp == 2:
            calc_mean = RC_stats_module.mean(numbers)
            print("The mean is", f"{calc_mean:.2f}", "\n")
        #Variance calculation
        elif user_inp == 3:
            calc_var = RC_stats_module.variance(numbers)
            print("The variance is", f"{calc_var:.2f}", "\n")
        #Standard deviation calculation
        elif user_inp == 4:
            calc_SD = RC_stats_module.std_dev(numbers)
            print("The standard deviation is", f"{calc_SD:.2f}", "\n")
        #Standard error calculation
        elif user_inp == 5:
            calc_SE = RC_stats_module.std_err(numbers)
            print("The standard error is", f"{calc_SE:.2f}", "\n")
        #Z-score calculation
        elif user_inp == 6:
            z_obs = int(input("Input an observation: "))
            calc_z = RC_stats_module.z_score(z_obs, numbers)
            print(z_obs, "is", f"{calc_z:.2f}", "standard deviations away from the mean\n")
        #Summary statistics
        elif user_inp == 7:
            calc_mean = RC_stats_module.mean(numbers)
            print("The mean is", f"{calc_mean:.2f}")
            calc_var = RC_stats_module.variance(numbers)
            print("The variance is", f"{calc_var:.2f}")
            calc_SD = RC_stats_module.std_dev(numbers)
            print("The standard deviation is", f"{calc_SD:.2f}")
            calc_SE = RC_stats_module.std_err(numbers)
            print("The standard error is", f"{calc_SE:.2f}", "\n")
```
